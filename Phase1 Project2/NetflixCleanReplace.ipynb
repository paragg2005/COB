## Checking for Outliers

```python

given_df['release_year'] = pd.to_numeric(given_df['release_year'], errors='coerce')


def handling_outliers(column):
    q1 = column.quantile(0.30)
    q2 = column.quantile(0.70)
    IQR = q2-q1
    lowerbound = q1-1.5*IQR
    upperbound = q2+1.5*IQR
    return column.clip(lower=lowerbound, upper=upperbound)
```

```python
given_df.loc[:,('release_year')]=handling_outliers(given_df['release_year'])


    
```

```python
given_df['director'].fillna("Unknown",inplace = True)
given_df['country'].fillna("Unknown",inplace = True)
```
